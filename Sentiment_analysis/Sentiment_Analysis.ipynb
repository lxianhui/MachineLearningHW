```python
## Preprocess Amazon dataset
new_amazon=[]
for i in amazon:
    ## Lowercase all of the words
    i[0] = i[0].lower()
    
    ## strip punctuation
    i[0] = i[0].translate(str.maketrans('','',string.punctuation))
    i[0] = i[0].translate(str.maketrans('','','1234567890'))
    
    ## strip stop words
    parse = i[0].split()
    parse = [word for word in parse if word not in stopwords.words('english')]
    
    ## Lemmatization of all the words
    ##lmtzr = WordNetLemmatizer()
    ##parse = [lmtzr.lemmatize(a) for a in parse]
    st = LancasterStemmer()
    parse = [st.stem(a) for a in parse]
    
    new_amazon.append([parse, i[-1]])
```

### Sentiment prediction: Naive Bayes model

```python
def normpdf(x, mean, sd):
    '''get pdf of x with mean and sd from a normal distribution'''
    if sd==0:
        if x == mean:
            return 1
        else: 
            return 0
    var = float(sd)**2
    pi = 3.1415926
    denom = (2*pi*var)**.5
    num = math.exp(-(float(x)-float(mean))**2/(2*var))
    return num/denom
```

```python
import numpy as np
from scipy.stats import norm

def Naive_Bayes(train, label):
    '''train the Naive Bayes model with training data and label, and return the [mean, sd] of each word given 
    zero or one'''
    df = DataFrame(train)
    df = df.transpose()
    p_word_given_1 = []
    p_word_given_0 = []
    for i in range(len(df.index)):
        c_word_given1 = []
        c_word_given0 = []
        for j in range(len(df.columns)):
            if label[j] == 1:
                c_word_given1.append(df.at[i, j])
            else:
                c_word_given0.append(df.at[i, j])
        mean1, sd1 = norm.fit(c_word_given1)
        mean0, sd0 = norm.fit(c_word_given0)
        p_word_given_1.append([mean1, sd1])
        p_word_given_0.append([mean0, sd0])
    return p_word_given_1, p_word_given_0
```

```python
import math
def predict(data, p_word_given_1, p_word_given_0):
    '''return the prediction (either 1 or 0) using Naive Bayes rule'''
    prob1 = 0
    prob0 = 0
    for i in range(len(data)):
        if data[i] == 0:
            continue
        else:
            p0 = normpdf(data[i], p_word_given_0[i][0], p_word_given_0[i][1])
            p1 = normpdf(data[i], p_word_given_1[i][0], p_word_given_1[i][1])
            p1 = math.log(p1+1)
            p0 = math.log(p0+1)
            prob1 = prob1 + p1
            ## print("prob1",prob1)
            prob0 = prob0 + p0
            ## print("prob0",prob0)
    ##print("prob1 = ", prob1, "prob0 = ", prob0)
    if prob1 > prob0:
        return 1
    else:
        return 0
```
